### Problem 4: Matrix Transpose and Inverse

```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)

np.random.seed(0)

# Creating 3 random 3 x 3 matrices with integers from -10 to 10
A1 = np.random.randint(-10, 11, (3, 3))
A2 = np.random.randint(-10, 11, (3, 3))
A3 = np.random.randint(-10, 11, (3, 3))

# Printing the matrices
print("Matrix A1:\n", A1)
print("\nMatrix A2:\n", A2)
print("\nMatrix A3:\n", A3)
```

```python
matrices = [A1, A2, A3]
inverse_matrices = []

# Looping through each matrix
for i, A in enumerate(matrices, start=1):

    # Computing the determinant to check if the inverse exists
    if abs(np.linalg.det(A)) < 1e-9:
        print("Inverse does not exist")
        inverse_matrices.append(None)
        continue

    # Compute the inverse using NumPy
    A_inv = np.linalg.inv(A)
    inverse_matrices.append(A_inv)
    print(f"Inverse Matrix A{i}:")
    print(A_inv)
```

```python
for i in range(len(matrices)):

    A = matrices[i]
    A_inv = inverse_matrices[i]

    if A_inv is None:
        print(f"A{i+1} has no inverse, cannot compute product.\n")
        continue

    product = A @ A_inv
    
    print(f"Identity Matrix A{i+1}:")
    print(np.abs(product))
```
